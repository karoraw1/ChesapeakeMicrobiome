```r
model_types = list('categorical' = c('cruise_id', 'StatName', 'Habitat'),
                   'continuous' = c('WTEMP', 'SALINITY', 'PO4F', 'LinearTime', 'NO2F','DO'))

cons_ord_res = data.frame('adjr2'=c(), 'pval'=c(), 'incremental.adjR2'=c(), 
                          'variable_selection_step' = c(), 'model.type'=c())
cntrx = 0

for (mt in names(model_types)){
    vars_mt = model_types[[mt]]
    for (v_i in 1:length(vars_mt)){
        cntrx = cntrx + 1
        cap.env.ind <- capscale(wunifrac_dist2 ~ 1 + . , data=as.data.frame(ENV.mini.2[,c(vars_mt[v_i])]))
        cap.env.inc <- capscale(wunifrac_dist2 ~ 1 + ., data=as.data.frame(ENV.mini.2[,vars_mt[1:v_i]]))
        cons_ord_res[cntrx, 'adjr2'] = round(RsquareAdj(cap.env.ind)$adj.r.squared, 3)
        cons_ord_res[cntrx, 'pval'] = anova(cap.env.ind)["Pr(>F)"]['Model',]
        cons_ord_res[cntrx, 'incremental.adjR2'] = round(RsquareAdj(cap.env.inc)$adj.r.squared, 3)
        cons_ord_res[cntrx, 'variable_selection_step'] = v_i
        cons_ord_res[cntrx, 'model.type'] = mt
        print(paste(cntrx, vars_mt[v_i]))
        flush.console()
    }
}


rownames(cons_ord_res) <- c(model_types[['categorical']], model_types[['continuous']])
cons_ord_res
write.table(cons_ord_res, file='/Volumes/KeithSSD/ChesapeakeMicrobiome/tables/constrainedOrdinationResults.txt',
           sep='\t', row.names=T)
```

```r

# package necessary items for summary
enrichment_pile = list('abunds'=prepro$feature_table,
                       'meta'=prepro$meta_data,
                       'std'=res[['out']])

# take the top scores from the standard 
std_i = enrichment_pile[['std']]
std_i_srt = std_i[order(std_i[,'W'], decreasing=T),]
# filter by effect size
effective_ids = eff_size[which(abs(eff_size$x) > 1.5), 'taxa_id']
bool1 = which(std_i_srt[,'W'] > nrow(prepro$feature_table)*0.6)
bool2 = which(std_i_srt$taxa_id %in% effective_ids)
bool3 = which(std_i_srt$taxa_id %in% c('nitrogen_fixation', 'human_pathogens_all',
                                       'aromatic_compound_degradation', 'photoautotrophy',
                                      'aerobic_ammonia_oxidation'))
bool3 = union(bool3, grep('Nitrosopumilaceae', std_i_srt$taxa_id))
std_i_sig = std_i_srt[union(intersect(bool1, bool2), bool3),]
std_i_sig[,'data.type'] <- NA
std_i_sig[which(std_i_sig$taxa_id %in% rownames(taxa_abund)), 'data.type'] = 'Family'
std_i_sig[which(std_i_sig$taxa_id %in% rownames(fxn_abund)), 'data.type'] = 'Function'
std_i_sig[which(std_i_sig$taxa_id %in% rownames(taxa_abund2)), 'data.type'] = 'Order'

duplicitous_upper_levels = c('Gammaproteobacteria.Thiomicrospirales', 'anoxygenic_photoautotrophy',
                             'Verrucomicrobiae.Verrucomicrobiales', 'dark_sulfide_oxidation',
                             'Verrucomicrobiae.Chthoniobacterales', 'nitrate_respiration',
                             'Deltaproteobacteria.Desulfobacterales',
                             'Phycisphaerae.Phycisphaerales',
                             'Bacteroidia.Cytophagales',
                             'Actinobacteria.Frankiales',
                             'Planctomycetacia.Planctomycetales',
                             'Gammaproteobacteria.Ectothiorhodospirales', 
                             'nitrogen_respiration',
                             'respiration_of_sulfur_compounds',
                             'Alphaproteobacteria.Caulobacterales',
                             'Actinobacteria.Micrococcales', 
                             'Oxyphotobacteria.Synechococcales',
                             'Deltaproteobacteria.Oligoflexales')

std_i_sigi = std_i_sig[which(!(std_i_sig$taxa_id %in% duplicitous_upper_levels)),]
std_top_scores = std_i_sigi[, c('taxa_id', 'W', 'data.type')]
colnames(std_top_scores) <- c('taxa_id', 'W.standard', 'data.type')
top_scores = std_top_scores
top_scores[which(is.na(top_scores$W.standard)), 'W.standard' ] = round(mean(top_scores$W.standard, na.rm=T))
top_scores = top_scores[which(!is.na(top_scores$W.standard)),]
rownames(top_scores) <- top_scores[,'taxa_id']
top_scores = top_scores[which(!(rownames(top_scores) %in% c('nitrification'))),]
top_scores = top_scores[order(top_scores$W.standard, decreasing=T),]

# pull out preprocessed metadta and abundances
ft_i = enrichment_pile[['abunds']]
ft_i = ft_i / rowMeans(ft_i)
md_i = enrichment_pile[['meta']]

# get aggregate abundances per cluster level
for (lev in levels(md_i[,'mclusts2'])){
    samps_i = md_i[which(md_i[,'mclusts2'] == lev), 'Sample.ID']
    aggd_abunds_i = rowMeans(ft_i[rownames(top_scores), samps_i])
    newName = as.character(lev)
    top_scores = cbind.data.frame(top_scores,
                                   data.frame(newName = aggd_abunds_i))
    colnames(top_scores)[colnames(top_scores) == 'newName'] = newName
}


top_scores[order(top_scores$data.type, top_scores$W.standard),]
min(apply(top_scores[,4:6], 1, max))
```
