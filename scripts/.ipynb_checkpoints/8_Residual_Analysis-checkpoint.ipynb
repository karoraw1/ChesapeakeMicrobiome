```r
numCores <- 3
counter = 0
folds = 10

results_objects = list()
for (d in names(data_objects)){
    predictor_list = colnames(data_objects[[d]])
    all_summaries = c(paste(d, 'Rsq', sep="."), paste(d, '(Intercept)', sep="."), predictor_list)
    vs_trial_df = data.frame(matrix(nrow=ncol(env_responses3), ncol=length(all_summaries)))
    rownames(vs_trial_df) = colnames(env_responses3)
    colnames(vs_trial_df) = all_summaries
    results_objects[[d]] = cbind(vs_trial_df)
    print(paste(dim(vs_trial_df), d))
}

```

```r
library(caret)
resid_res_file = '/Volumes/KeithSSD/ChesapeakeMicrobiome/data/residual_analysis/responsive_populations_with.OEU.RData'

if (file.exists(resid_res_file)) {
    load(resid_res_file)
    print('Loaded residual results')
} else {
    suppressPackageStartupMessages(library(glmnet))
    suppressPackageStartupMessages(library(doParallel))
    registerDoParallel(3)
    start_time = Sys.time()
    
    for (trial in names(data_objects)){
        # extract predictor matrix
        predictors.t = as.matrix(data_objects[[trial]])
        pred_cols = colnames(predictors.t)
        print(c("Trial", trial))
        
        # iterate over base variables
        for (resp in fullSpectrum){
            
            # set seed by counter
            counter = counter + 1
            
            # modify into high and low end column names
            p.t = paste(resp, 'Pos', sep=".")
            n.t = paste(resp, 'Neg', sep=".")
            
            # pull actual response data and name
            response.p.t = env_responses3[,p.t]
            response.n.t = env_responses3[,n.t]
            names(response.n.t) <- rownames(env_responses3)
            names(response.p.t) <- rownames(env_responses3)
            
            for (split.i in 1:folds) {
                
                this_counter = counter + split.i
                set.seed(this_counter)
                # sample to get train indexes 
                train_idx = createDataPartition(response.n.t, p=.75)[['Resample1']]
                # differentiate test indexes
                test_idx = setdiff(1:nrow(predictors.t), train_idx)
                # check they sum 
                if ((length(test_idx) + length(train_idx) )!= nrow(predictors.t)){
                    cow
                }
                # split predictor set
                predictors.t.tr <- predictors.t[train_idx, ]; 
                predictors.t.tst <- predictors.t[test_idx, ]; 
                
                # split response sets explicitly
                response.p.t.tr <- response.p.t[train_idx]
                response.p.t.tst <- response.p.t[test_idx]
                
                response.n.t.tr <- response.n.t[train_idx]
                response.n.t.tst <- response.n.t[test_idx]
                
                cvfit.p.tr <- cv.glmnet(x=predictors.t.tr, y=response.p.t.tr, family = "gaussian", 
                                        nfolds = 5, parallel = F, intercept=F, alpha=1,
                                        lower.limits=c(-Inf, rep(0, ncol(predictors.t.tr))))
                
                cvfit.n.tr <- cv.glmnet(x=predictors.t.tr, y=response.n.t.tr, family = "gaussian", 
                                        nfolds = 5, parallel = F, intercept=F, alpha=1,
                                        lower.limits=c(-Inf, rep(0, ncol(predictors.t.tr))))
                                                     
                # predict on held out (pos) data
                fitted_values.p.tst = predict(cvfit.p.tr, newx=predictors.t.tst, s = 'lambda.min')
                rsq.p.tst = postResample(fitted_values.p.tst, response.p.t.tst)['Rsquared']
                rsq.p.tst = matrix(rsq.p.tst, nrow=1,ncol=1, 
                                 dimnames=list(c(paste(trial,'Rsq',sep=".")), c(split.i)))
                coef.p.tr = as.matrix(coef(cvfit.p.tr, s = "lambda.min"))
                colnames(coef.p.tr) <- c(split.i)
                
                # predict on held out (neg) data
                fitted_values.n.tst = predict(cvfit.n.tr, newx=predictors.t.tst, s = 'lambda.min')
                rsq.n.tst = postResample(fitted_values.n.tst, response.n.t.tst)['Rsquared']
                rsq.n.tst = matrix(rsq.n.tst, nrow=1,ncol=1, 
                                 dimnames=list(c(paste(trial,'Rsq',sep=".")), c(split.i)))
                coef.n.tr = as.matrix(coef(cvfit.n.tr, s = "lambda.min"))
                colnames(coef.n.tr) <- c(split.i)
                
                if (split.i == 1) {
                    fold.p.scores = rbind(coef.p.tr, rsq.p.tst)
                    fold.n.scores = rbind(coef.n.tr, rsq.n.tst)
                } else {
                    fold.p.scores = cbind(fold.p.scores, rbind(coef.p.tr, rsq.p.tst))
                    fold.n.scores = cbind(fold.n.scores, rbind(coef.n.tr, rsq.n.tst))
                }
            }
            
            # move counter forward to initiate new seeds 
            counter = counter + folds
            # fold the scores
            folded.p.scores = apply(fold.p.scores, 1, function(x) {mean(x, na.rm=T)})
            na_count.p = sum((colSums(is.na(fold.p.scores))))
            # give the intercept a model specific name
            names(folded.p.scores)[names(folded.p.scores) == '(Intercept)'] = paste(trial, '(Intercept)', sep=".")
            results_objects[[trial]][p.t, names(folded.p.scores)] = folded.p.scores
            
            folded.n.scores = apply(fold.n.scores, 1, function(x) {mean(x, na.rm=T)})
            na_count.n = sum((colSums(is.na(fold.n.scores))))
            # give the intercept a model specific name
            names(folded.n.scores)[names(folded.n.scores) == '(Intercept)'] = paste(trial, '(Intercept)', sep=".")
            
            results_objects[[trial]][n.t, names(folded.n.scores)] = folded.n.scores

            progress_meter = counter/(length(fullSpectrum)*(folds+1)*length(data_objects))
            end_time = Sys.time();
            print(paste('Completed: ', round(progress_meter*100, 1), "%, Nulls (P): ", na_count.p, 
                                                                     "%, Nulls (N): ", na_count.n, sep="" ))
            print(end_time - start_time);
            flush.console()
        }
    }
    stopImplicitCluster()
    save(results_objects, file=resid_res_file)
    print('Saved residual results')
}
```

```r
# scoring O2 regression 
# "aerobic" ASVs with positive coefficients for DO.Pos = TP
# "anaerobic" ASVs with positive coefficients for DO.Pos = TN
# "aerobic" ASVs with positive coefficients for DO.Neg = FN
# "anaerobic" ASVs with positive coefficients for DO.Neg = TN

extract_tabled_features <- function (resp, thresh, feat_data){
    pops = list()
    cols_ = colnames(data_objects[[feat_data]])
    df = results_objects[[feat_data]]
    #print(paste(feat_data, ": ", length(cols_), sep=""))
    allp_row = df[paste(resp, 'Pos', sep="."), cols_]
    p.pos_row = allp_row[which(allp_row > thresh)]
    alln_row = df[paste(resp, 'Neg', sep="."), cols_]
    n.pos_row = alln_row[which(alln_row > thresh)] 
    #print(paste(resp, ": (Pos, ", length(p.pos_row), ") | (Neg, ", length(n.pos_row),")", sep=""))
    pops = list('pos'=p.pos_row, 'neg'=n.pos_row)
    pops
}

# the extracted populations, the nutrient, the predictor type 
confuseNut <- function(ex_ta_fts, nutType, data_type){
    if (data_type == 'ASV.OEU') {
        ex_ta_fts2 = list('pos' = c(), 'neg'=c())
        for (n in names(ex_ta_fts$pos)) {
            t.per.e = length(final_groups[[n]])
            coeff = ex_ta_fts$pos[n]/t.per.e
            all_names = c(names(ex_ta_fts2$pos), final_groups[[n]])
            ex_ta_fts2$pos = c(ex_ta_fts2$pos, rep(coeff, t.per.e))
            names(ex_ta_fts2$pos) <- all_names
        }
        for (n in names(ex_ta_fts$neg)) {
            t.per.e = length(final_groups[[n]])
            coeff = ex_ta_fts$neg[n]/t.per.e
            all_names = c(names(ex_ta_fts2$neg), final_groups[[n]])
            ex_ta_fts2$neg = c(ex_ta_fts2$neg, rep(coeff, t.per.e))
            names(ex_ta_fts2$neg) <- all_names
        }
        ex_ta_fts <- ex_ta_fts2
    }

    if (nutType %in% c('aerobic', 'anaerobic')){
        if (data_type == 'Fxn.SILVA'){ 
            aer_pool = aerobic_proc; anaer_pool = anaerobic_proc;
        } else {
            aer_pool = otus_per_function[['aerobic']]; 
            anaer_pool = otus_per_function[['anaerobic']];
        }
        TP.o2 = sum((names(ex_ta_fts[['pos']]) %in% aer_pool))
        FP.o2 = sum((names(ex_ta_fts[['pos']]) %in% anaer_pool))
        FN.o2 = sum((names(ex_ta_fts[['neg']]) %in% aer_pool))
        TN.o2 = sum((names(ex_ta_fts[['neg']]) %in% anaer_pool))
        if (nutType == 'aerobic'){
            return((TP.o2)/(TP.o2 + FP.o2))
        } else {
            return((TN.o2)/(TN.o2 + FP.o2))
        }
    } else if (nutType == 'NO2F') {
        if (data_type == 'Fxn.SILVA'){ 
            plus_pool = plus_nitro; minus_pool = minus_nitro;
         } else {
            plus_pool = otus_per_function[['plus_nitro']]; 
            minus_pool = otus_per_function[['minus_nitro']];
        }        
        TP.no2 = ex_ta_fts[['pos']]; TP.no2 = names(TP.no2)[(names(TP.no2) %in% plus_pool)]
        FP.no2 = ex_ta_fts[['pos']]; FP.no2 = names(FP.no2)[(names(FP.no2) %in% minus_pool)]
        TN.no2 = ex_ta_fts[['neg']]; TN.no2 = names(TN.no2)[(names(TN.no2) %in% minus_pool)]
        FN.no2 = ex_ta_fts[['neg']]; FN.no2 = names(FN.no2)[(names(FN.no2) %in% plus_pool)]
        return(length(TP.no2)/(length(TP.no2) + length(FN.no2)))
    } else if (nutType == 'PO4F') {
        if (data_type == 'Fxn.SILVA'){ 
            highp_pool = anaerobic_proc; lowp_pool = aerobic_proc;
        } else {
            plus_pool = otus_per_function[['anaerobic']]; 
            minus_pool = otus_per_function[['aerobic']];
        }        
        FN.po4 = sum((names(ex_ta_fts[['pos']]) %in% aerobic_proc))
        TP.po4 = sum((names(ex_ta_fts[['neg']]) %in% aerobic_proc))
        TN.po4 = sum((names(ex_ta_fts[['pos']]) %in% anaerobic_proc))
        FP.po4 = sum((names(ex_ta_fts[['neg']]) %in% anaerobic_proc))
        return(length(TP.po4)/(length(TP.po4) + length(FN.po4)))
    } else {
        cow
    }
}

merge_pops_identified <- function(raw_asvs, detr_asvs){
    crosspos = intersect(names(raw_asvs$pos), names(detr_asvs$pos))
    crossneg = intersect(names(raw_asvs$neg), names(detr_asvs$neg))
    cross_ids = list('pos'=((raw_asvs$pos[crosspos] + detr_asvs$pos[crosspos])/2) , 
                     'neg'=((raw_asvs$neg[crossneg] + detr_asvs$neg[crossneg])/2))
    return(cross_ids)
}
```

```r
oeu_to_constituents <- function(x) { 
    x2 = c()
    for (n in names(x)) {
        t.per.e = length(final_groups[[n]])
        coeff = x[n]/t.per.e
        all_names = c(names(x2), final_groups[[n]])
        x2 = c(x2, rep(coeff, t.per.e))
        names(x2) <- all_names
    }
    return(x2)
}

detrend.O2 = oeu_to_constituents(pop_box[['anaerobic']][['ASV.OEU']][['detrended']])
detrend.PO4 = oeu_to_constituents(pop_box[['PO4F']][['ASV.OEU']][['detrended']])
detrend.NO2 = oeu_to_constituents(pop_box[['NO2F']][['ASV.OEU']][['detrended']])

raw.O2 = oeu_to_constituents(pop_box[['anaerobic']][['ASV.OEU']][['raw']])
raw.PO4 = oeu_to_constituents(pop_box[['PO4F']][['ASV.OEU']][['raw']])
raw.NO2 = oeu_to_constituents(pop_box[['NO2F']][['ASV.OEU']][['raw']])


RelA_unrar_df = unrar_df / rowSums(unrar_df)
```

```r
abund_mult1 <- function(lset, trial){
    lsetnew = c()
    for (n in names(lset)){lsetnew[n] = mean(data_objects[[trial]][,n])*lset[n]}
    lsetnew
}

trends_n_vars = c('Latitude', 'Seasonality', 'DO', 'DO_residuals', 'NO2F', 'NO2F_residuals', 'PO4F', 
                  'PO4F_residuals')

reordered_results = list()
for (trial in names(data_objects)[1:3]){
    microbes = colnames(data_objects[[trial]])
    counter = 1
    comp_plot_sets = data.frame('Direction'=c(), 'Variable'=c(), 'OTU'=c(), 'Value'=c())
    for (factor in trends_n_vars){
        coefs_ = t(vs_trial_df[factor, microbes])
        valsets = list('Pos' = base::sort(abund_mult1(coefs_[(coefs_[,1] > 0),], trial), decreasing=T), 
                       'Neg' = base::sort(abund_mult1(coefs_[(coefs_[,1] < 0),], trial), decreasing=F))
        
        valsets[['Pos']] = valsets[['Pos']] / sum(valsets[['Pos']])
        valsets[['Neg']] = valsets[['Neg']] / sum(valsets[['Neg']])
        
        for (n_v in names(valsets)){
            for (otu_h in names(valsets[[n_v]])){
                comp_plot_sets[counter, 'Direction'] = n_v
                comp_plot_sets[counter, 'Variable'] = factor
                comp_plot_sets[counter, 'OTU'] = otu_h
                comp_plot_sets[counter, 'Value'] = valsets[[n_v]][otu_h]
                counter = counter + 1
            }
        }
        
        message(paste(trial, ": ", factor, " +", length(valsets[['Pos']]), " -", length(valsets[['Neg']]), sep=""))
        
        for (rank_ in 1:15){
            print(paste(factor, 'POS', rank_, names(valsets[['Pos']])[rank_], round(valsets[['Pos']][rank_], 5), sep="    "))
        }
        print(paste(paste(rep('0', 15), rep('1', 15), sep=""), collapse=""))
        
        for (rank_ in 1:15){
            print(paste(factor, 'NEG', rank_, names(valsets[['Neg']])[rank_], round(valsets[['Neg']][rank_], 5), sep="    "))
        }
        print(paste(paste(rep('*', 30), rep('-', 30), sep=""), collapse=""))
    }
    
    comp_plot_sets[,'Variable'] = factor(comp_plot_sets[,'Variable'])
    comp_plot_sets[,'Direction'] = factor(comp_plot_sets[,'Direction'])
    reordered_results[[trial]] = comp_plot_sets
}

reordered_results[['Raw.ASVs']] = merge(reordered_results[['Raw.ASVs']], taxa_df, by.x="OTU", by.y="row.names")
```

```r
# addalpha()
addalpha <- function(colors, alpha=0.6) {
  r <- col2rgb(colors, alpha=T)
  # Apply alpha
  r[4,] <- alpha*255
  r <- r/255.0
  return(rgb(r[1,], r[2,], r[3,], r[4,]))
}

# colorRampPaletteAlpha()
colorRampPaletteAlpha <- function(colors, n=32, interpolate='linear') {
  # Create the color ramp normally
  cr <- colorRampPalette(colors, interpolate=interpolate)(n)
  # Find the alpha channel
  a <- col2rgb(colors, alpha=T)[4,]
  # Interpolate
        if (interpolate=='linear') {
            l <- approx(a, n=n)
        } else {
            l <- spline(a, n=n)
        }
        l$y[l$y > 255] <- 255 # Clamp if spline is > 255
        cr <- addalpha(cr, l$y/255.0)
        return(cr)
}
```
