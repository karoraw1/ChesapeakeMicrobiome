In this block we rename the columns, rename the stations, and convert the datestrings to pandas datetime timestamps

````python
import pandas as pd
import numpy as np
import os

env_data_dir = "/Volumes/KeithSSD/ChesapeakeMicrobiome/data/environmental_raw_data"
noaa_dir = os.path.join(env_data_dir, 'NOAA_ISD_ClimateData')
wind_file = os.path.join(noaa_dir, "NOAA_ISD_ClimateData_MarylandStations.txt.bz2")

climate_df = pd.read_csv(wind_file, sep='\t', header=None, compression='infer')
header_line = climate_df.iloc[0, 0].split()
data_matrix = list(map(lambda x: climate_df.iloc[x, 0].split(), range(1, len(climate_df))))
climate_df = pd.DataFrame(data_matrix, columns = header_line)

# keep columns we might need and give meaningful names
newcol_names = {'USAF':'STATION_NUMBER',  
                "YR--MODAHRMN":'DATE_TIME',
                'SKC': 'SKY_COVER',      # dimensionless, categorical? 
                'VSB': 'VISIBILITY',     # IN STATUTE MILES TO NEAREST TENTH
                'DIR': 'WIND_DIRECTION', # IN COMPASS DEGREES, 990 = VARIABLE, REPORTED AS
                'SPD': 'WIND_SPEED',     # IN MILES PER HOUR
                'PCP24': 'PRECIP24HR'}

climate_dfx = climate_df[newcol_names.keys()]
climate_dfx = climate_dfx.rename(columns=newcol_names)

# rename the bases to human names and parse dates
# date format: YEAR-MONTH-DAY-HOUR-MINUTE IN GREENWICH MEAN TIME (GMT)

base_names = {'720384':'BayBridge', '724060': 'BWI', '745940': 'Andrews'}
fix_date_fmt = lambda x: x[4:6] + "/" + x[6:8] + "/" + x[:4] + " " + x[8:10] + ":" + x[10:]

climate_dfx['DATE_TIME'] = pd.to_datetime(climate_dfx.DATE_TIME.apply(fix_date_fmt), infer_datetime_format=True)
climate_dfx['STATION_NUMBER'] = climate_dfx.STATION_NUMBER.map(base_names)
print("Weather stations collected:", climate_dfx['STATION_NUMBER'].unique())
print("Weather date start:", climate_dfx['DATE_TIME'].min())
print("Weather date end:", climate_dfx['DATE_TIME'].max())
print("Weather data types:", list(climate_dfx.columns)[2:])
````

#### Here we manually replace variable length strings of asterisks with NaNs, convert the Sky Cover variable into categorical with normal names and order, and conver the wind direction into a categorical of cardinal directions. 

#### This data is a mix of types and recorded and irregular intervals, so we write a function to aggregate into daily intervals

````python
# clean up the sky cover column
climate_df2 = climate_dfx.copy()
ordered_renamer = [("***", np.nan), ('CLR','Clear'), ('SCT','Scattered'), ('BKN','Broken'), ('OVC','Overcast'), ('POB','PartiallyObscured'), ('OBS','Obscured')]
skycov_dict = {i[0]:i[1] for i in ordered_renamer}
skycov_lvls = [i[1] for i in ordered_renamer[1:]]
climate_df2['SKY_COVER'] = climate_df2.SKY_COVER.map(skycov_dict)
climate_df2['SKY_COVER'] = pd.Categorical(climate_df2['SKY_COVER'], categories=skycov_lvls, ordered=True)
print("Sky cover category counts:")
skycov_catcnts = np.unique(climate_df2['SKY_COVER'].dropna(), return_counts=True)
print([(i, j) for i, j in zip(skycov_catcnts[0], skycov_catcnts[1])])

# clean up other NaNs
nan_strings = {'VISIBILITY':'****', 'WIND_DIRECTION': '***', 'WIND_SPEED':'***'}
for col_, str_nan in nan_strings.items():
    climate_df2[col_] = climate_df2[col_].map(lambda x: np.nan if x == str_nan else x)
    if col_ in ['VISIBILITY', 'WIND_SPEED']:
        climate_df2[col_] = climate_df2[col_].astype(np.float64)

climate_df2['PRECIP24HR'] = climate_df2['PRECIP24HR'].astype(str).apply(lambda x: x if (not "*" in x) and (not "T" in x) and (not 'No' in x) else np.nan)
climate_df2['PRECIP24HR'] = climate_df2['PRECIP24HR'].astype(np.float64)


nullcount = climate_df2[['STATION_NUMBER']].merge(climate_df2.iloc[:, 2:].isnull(), left_index=True, right_index=True, validate='1:1')
print("Null counts by stations:")
print(nullcount.groupby('STATION_NUMBER').agg(np.sum))

# convert compass degress into direction categorical 
wind_direction = {"NNE": (11.25, 33.75), "NE": (33.75, 56.25),
                  "ENE": (56.25, 78.75), "E": (78.75, 101.25), "ESE": (101.25, 123.75),
                  "SE": (123.75, 146.25), "SSE": (146.25, 168.75), "S": (168.75, 191.25),
                  "SSW": (191.25, 213.75), "SW": (213.75, 236.25), "WSW": (236.25, 258.75),
                  "W": (258.75, 281.25), "WNW": (281.25, 303.75), "NW": (303.75, 326.25),
                  "NNW": (326.25, 348.75)}

def map_directionality(compass_dir):
    if np.isnan(compass_dir):
        return np.nan
    if (compass_dir >= 348.75) or (compass_dir < 11.25):
        return "N"
    for wd, (l1, l2) in wind_direction.items():
        if (compass_dir >= l1) & (compass_dir < l2):
            return wd
    raise ValueError('{} is not wind direction?'.format(compass_dir))

climate_df2['WIND_CARDINAL'] = climate_df2['WIND_DIRECTION'].astype(np.float64).apply(map_directionality)
# split the data by station
subclimes = {i: climate_df2.copy()[climate_df2.STATION_NUMBER == i] for i in climate_df2.STATION_NUMBER.unique()}

# create a measurment interval variable for each (will use below)
def add_interval(subclime_):
    late_time = subclime_['DATE_TIME'].iloc[1:].values
    early_time = subclime_['DATE_TIME'].iloc[:-1].values
    subtract_time = lambda x, y: pd.Timedelta(x - y).seconds / 60.0
    time_diff = np.array([subtract_time(x, y) for x, y in zip(late_time, early_time)])
    print("Average {}, Std {}, Nulls {}".format(time_diff.mean(), time_diff.std(), np.isnan(time_diff).sum()))
    time_diff = np.array([time_diff.mean().round()] + list(time_diff))
    return pd.Series(time_diff, index=subclime_.index)

for stat_, stat_df in subclimes.items():
    stat_df['TIME_INTERVAL'] = add_interval(stat_df)
    stat_df['WIND_CARDINAL'] = stat_df[['WIND_CARDINAL', 'TIME_INTERVAL']].apply(tuple, axis=1)
    subclimes[stat_] = stat_df.copy()

# this data set has null values for 17% of the time covered
del subclimes['BayBridge']
    
# define a function that aggregates the directional data by 
# counting the fraction of intervals with "NE" or "SW" in 
# a time step. and all other columns needed

def aggregate_af_data(x, card_dir):
    if card_dir == 'vis':
        return x.dropna().mean()
    elif card_dir =='rain':
        return np.nansum(x)
    elif card_dir == 'sky':
        x = x.dropna()
        if len(x) > 0:
            return x.mode().values[0]
        else:
            return np.nan
    elif card_dir == 'time':
        return x.sum()
    elif (card_dir in wind_direction.keys()) or card_dir == 'N':
        filt_x = [j for (i, j) in list(x) if i == card_dir]
        return sum(filt_x)
    else:
        raise ValueError('{} not an option'.format(card_dir))

# define function to apply that function across columns and get out a new dataframe        
def frequency_aggregator(subclime_, freq_='1d'):
    wind_group = pd.Grouper(key='DATE_TIME', freq=freq_)
    col_cardDir = [('WIND_CARDINAL', 'NE'), ('WIND_CARDINAL', 'SW'), ('PRECIP24HR', 'rain'),
                   ('VISIBILITY', 'vis'), ('TIME_INTERVAL', 'time'), ('SKY_COVER', 'sky')]
    aggd_cols = []
    for col_, cd_ in col_cardDir:
        sd_sub = subclime_[['DATE_TIME', col_]].copy()
        sdsub_agg = sd_sub.groupby(wind_group).agg(lambda x: aggregate_af_data(x, cd_))
        sdsub_agg = sdsub_agg[sdsub_agg.columns[0]]
        if col_ == 'WIND_CARDINAL':
            sdsub_agg.name = col_ + "_" + cd_
        elif col_ == 'SKY_COVER':
            sdsub_agg = sdsub_agg.astype('category')
            new_cats = [i for i in sd_sub.SKY_COVER.cat.categories if i in list(sdsub_agg.cat.categories)]
            sdsub_agg = sdsub_agg.cat.reorder_categories(new_cats)        
        aggd_cols.append(sdsub_agg.copy())
    
    subclime_agg = pd.concat(aggd_cols, axis=1, verify_integrity=True)
    return subclime_agg

# aggregate into daily averages for the background 
for stat_, stat_df in subclimes.items():
    subclimes[stat_] = frequency_aggregator(stat_df.copy())
````

#### The unit of measurement collection of the microbial data is cruises. The boat goes down the bay over multiple days, sometimes seperated by a week or so, and hits all the stations for samples. I manually entered the time between cruises and then number them in a variable for aggregation later and easier plotting. 

````python
# and "cruisly averages" in the main
# present as the average of all both stations so as to include error bars

cruise_bounds = ["01/01/2016", "02/01/2016", "03/01/2016", "04/01/2016", "05/01/2016", "06/01/2016",
                 "06/15/2016", "07/01/2016", "07/15/2016", "08/01/2016", "08/15/2016", "09/01/2016", 
                 "10/01/2016", "11/01/2016", "12/01/2016", "01/01/2017", "02/03/2017", "03/01/2017", 
                 "04/01/2017", "05/01/2017", "06/01/2017", "06/15/2017", "07/01/2017", "07/18/2017", 
                 "08/01/2017", "08/15/2017", "09/06/2017", "10/01/2017"]

cb_limits = list(zip(cruise_bounds[:-1], cruise_bounds[1:]))

def add_cruise_id(dbt, date_col):
    # takes the dataframe and makes a new column
    dbt['cruise_id'] = pd.Series(index=dbt.index)
    # turns the bounds into datetime objects
    cruise_bounds_dt = [pd.to_datetime(cb) for cb in cruise_bounds]
    cb_limits2 = list(zip(cruise_bounds_dt[:-1], cruise_bounds_dt[1:]))
    
    # sometimes the date col is in the index sometimes its a column
    if dbt.index.name == date_col:
        dbt = dbt.reset_index().copy()
    assert date_col in list(dbt.columns)
    
    for t_ix, (tl, th) in enumerate(cb_limits2):
        time_range_bool = (dbt[date_col] >= tl) & (dbt[date_col] < th)
        dbt.loc[time_range_bool, 'cruise_id'] = t_ix
    
    dbt = dbt.copy().loc[dbt['cruise_id'].notnull(), :]
    dbt['cruise_id'] = pd.Categorical(list(dbt['cruise_id'].astype(int)), ordered=True, 
                                      categories=range(len(cb_limits2)))
    if 'Cruise' in list(dbt.columns):
        dbt.drop('Cruise', axis=1, inplace=True)
    
    return dbt.copy().sort_values(by=date_col)

for stat_n, stat_df in subclimes.items():
    subclimes[stat_n] = add_cruise_id(stat_df, 'DATE_TIME')

print("Dates per cruise for {}".format(stat_n))
dates_per_cruise = np.unique(subclimes[stat_n]['cruise_id'], return_counts=1)
print([(i, j) for i, j in zip(dates_per_cruise[0], dates_per_cruise[1])])
````

# Satellite Data

#### Now we read in the daily satellite files for PAR and CHLA in 4km resolution. We  cut out the portion covering the chesapeake bay and save it into dataframes for each data type which we store in a dictionary. This is a time and resource intensive process and so to avoid repeating it, we save the dictionary as a binary python readable file (a pickle)

````python
from netCDF4 import Dataset
import os; import pandas as pd;
import numpy as np
import gc
import pickle 


# load the file lists and split them
env_data_dir = "/Volumes/KeithSSD/ChesapeakeMicrobiome/data/environmental_raw_data"
nasa_dir = os.path.join(env_data_dir, "OceanColorData")
nasa_files = [os.path.join(nasa_dir, i) for i in os.listdir(nasa_dir) if i.endswith("nc")]
chlor_files = list(filter(lambda x: 'CHL_chlor_a' in x, nasa_files))
par_files = list(filter(lambda x: 'PAR_par' in x, nasa_files))
    
# check for pickle
map_data_file = os.path.join(env_data_dir, 'processed_par_chla_data.pickle')
if os.path.exists(map_data_file):
    with open( map_data_file, "rb" ) as fh:
        map_data = pickle.load(fh)
else:
    # operate over each data type independendly
    map_data = {}
    for var_col, f_lst in zip(['par', 'chlor_a'], [par_files, chlor_files]):
        # create a dataframe for each data type
        nasa_cols = ['time_coverage_start', 'time_coverage_end', 'measurement', "unit"]
        nasa_df = pd.DataFrame(index=range(len(f_lst)), columns=nasa_cols)
        # go into each file 
        for df_ix in nasa_df.index:
            # open it
            nc_file = f_lst[df_ix]
            fh = Dataset(nc_file, mode='r')

            # pulls out the measurment time interval and units
            nasa_df.loc[df_ix, "time_coverage_start"] = pd.to_datetime(fh.time_coverage_start).tz_convert('EST')
            nasa_df.loc[df_ix, "time_coverage_end"] = pd.to_datetime(fh.time_coverage_end).tz_convert('EST')
            nasa_df.loc[df_ix, "unit"] = fh.variables[var_col].units

            # extracts the geoposition vectors and data matrix
            lons = fh.variables['lon'][:]
            lats = fh.variables['lat'][:]
            mapped_data = fh.variables[var_col][:]

            # find the start and end index numbers of your ROI
            bool_x = np.arange(0, lats.shape[0])[(lats <= 39.5) & (lats >= 37)]
            bool_y = np.arange(0, lons.shape[0])[(lons <= -76) & (lons >= -76.5)]

            # pull out the data from the matrix
            extracted = mapped_data[bool_x[0]:bool_x[-1]+1, bool_y[0]:bool_y[-1]+1]

            # remove mask without any invalid values
            unmasked = extracted[~extracted.mask].data

            # if less than a quarter of it is good data
            full_data_size = len(extracted.flatten())
            if len(unmasked) < (0.25*full_data_size):
                nasa_df.loc[df_ix, 'measurement'] = np.nan
            else:
                # store the average value over that region 
                nasa_df.loc[df_ix, 'measurement'] = unmasked.mean()
            # closes the file 
            fh.close()
            gc.collect()
            print(df_ix/len(nasa_df))

        map_data[var_col] = nasa_df.copy()
    if not os.path.exists(map_data_file):
        pickle.dump( map_data, open( map_data_file , "wb" ))
````

#### We read in the discharge data for the major sources of volume in the Chesapeake. They are highly correlated in their dynamics, so we only retain 2 of the largest ones for later analysis. It is given in 15 minute intervals which we scale down into daily means. 


````python
expected_columns = {"Discharge": ["146731_00060", "122065_00060", "69928_00060", "147046_00060", "69512_00060"]}
code2col = {val_i:key for key, val in expected_columns.items() for val_i in val}

# load the files
discharge_dir = os.path.join(env_data_dir, "USGS_Discharge")
discharge_files = [f for f in os.listdir(discharge_dir) if f.endswith(".txt.bz2")]
dischargeNamePath = [(f.split("_")[0], os.path.join(discharge_dir, f)) for f in discharge_files]

# read in rappahanok
df_list = []
for riv_name, riv_file in dischargeNamePath:
    df = pd.read_csv(riv_file, sep="\t", comment='#', header=0, low_memory=False).drop(0, axis=0)
    # get data quality code columns
    code_cols = [c for c in df.columns if c.endswith("_cd") and not c.startswith("agency") and not c.startswith("tz")]
    # remove any data not approved for publication
    df2 = df.copy()[((df.loc[:, code_cols] != 'A') & (df.loc[:, code_cols].notnull())).sum(1) == 0]
    df2.drop(code_cols, axis=1, inplace=True)
    df2.rename(mapper=code2col, axis=1, inplace=True)
    df2['Datetime'] = pd.to_datetime(df2['datetime'], format='%Y-%m-%d %H:%M:%S')
    df2.set_index('Datetime', inplace=True)
    df2 = df2[df2.index.notnull()]
    df2.drop(["agency_cd", "site_no", "datetime", "tz_cd"], axis=1, inplace=True)
    df2 = df2.astype(np.float64).loc[:, ['Discharge']]
    df2 = df2.groupby(pd.Grouper(freq='D')).agg(np.nanmean)#.dropna(how='any', axis=0)
    df2.columns= [i+"_"+riv_name for i in df2.columns]
    df_list.append(df2.copy())

        
mega_df = pd.concat(df_list, axis=1)

print("River stations collected:", mega_df.columns)
print("Discharge date start:", mega_df.index.min())
print("Discharge date end:", mega_df.index.max())
print("Discharge data types:", mega_df.dtypes)
print("Discharge missing data:\n", mega_df.isnull().sum())
````

Here we merge all these data sources, which are collected/provided in such a way that doesn't lend itself to localization with a particular station or depth. To do this, we modify their time-stamp indexes to match as time-agnostic daily intervals. We are also mindful to align them, removing time periods only present in one or another data set. 

````python
par_data = map_data['par']
par_data['DATE_TIME'] = pd.to_datetime(par_data['DATE_TIME'].dt.date)
par_data = par_data.set_index('DATE_TIME').asfreq('d')
par_data.columns = [i+"_PAR" for i in par_data.columns]

chlor_data = map_data['chlor_a']
chlor_data['DATE_TIME'] = pd.to_datetime(chlor_data['DATE_TIME'].dt.date)
chlor_data = chlor_data.set_index('DATE_TIME').asfreq('d')
chlor_data.columns = [i+"_CHLA" for i in chlor_data.columns]

andrews_wind = subclimes['Andrews'].set_index('DATE_TIME').asfreq('d')
bwi_wind = subclimes['BWI'].set_index('DATE_TIME').asfreq('d')
andrews_wind.columns = [i+"_ANDREWS" for i in andrews_wind.columns]
bwi_wind.columns = [i+"_BWI" for i in bwi_wind.columns]

baywide_datax = par_data.merge(chlor_data, how='inner', left_index =True, right_index=True, validate="1:1")
print("CHLOR+PAR:", baywide_datax.shape)
baywide_datax = baywide_datax.merge(andrews_wind, how='inner', left_index =True, right_index=True, validate="1:1")
print("+WIND:", baywide_datax.shape)
baywide_datax = baywide_datax.merge(bwi_wind, how='inner', left_index =True, right_index=True, validate="1:1")
print("+DISCHARGE:", baywide_datax.shape)
baywide_datax = baywide_datax.merge(mega_df, how='inner', left_index =True, right_index=True, validate="1:1")

print('all data date min:', baywide_datax.index.min())
print('all data date max:', baywide_datax.index.max())
print("Check the number cruise IDs match before dropping: ")
print((baywide_datax['cruise_id_PAR'] == baywide_datax['cruise_id_CHLA']).sum())
print((baywide_datax['cruise_id_ANDREWS'] == baywide_datax['cruise_id_BWI']).sum())
baywide_datax.drop(['cruise_id_PAR', 'cruise_id_CHLA', 'cruise_id_ANDREWS'], inplace=True, axis=1)

baywide_datax.rename(columns={'cruise_id_BWI': 'cruise_id'}, inplace=True)
````

Because daily directional wind duration is often zero, the resultant ratios will have an overabundance of NaN and infinite values because of division by zero issues. Here we use a 14 day centered rolling average instead to minimize that without significantly changing higher order aggregations. 

````python
baywide_data = baywide_datax.copy()
wind_vars = ['WIND_CARDINAL_NE_BWI', 'WIND_CARDINAL_NE_ANDREWS', 'WIND_CARDINAL_SW_BWI', 'WIND_CARDINAL_SW_ANDREWS']
for wv in wind_vars:
    baywide_data[wv] = baywide_data[wv].rolling(14, center=True).mean()
    print(wv, baywide_data[wv].isnull().sum())
    
baywide_data['WIND_RATIO_BWI'] = (baywide_data['WIND_CARDINAL_SW_BWI']) / (baywide_data['WIND_CARDINAL_NE_BWI'])
baywide_data['WIND_RATIO_ANDREWS'] = (baywide_data['WIND_CARDINAL_SW_ANDREWS']) / (baywide_data['WIND_CARDINAL_NE_ANDREWS'])
bwimax = np.percentile(baywide_data.loc[baywide_data['WIND_RATIO_BWI'] != np.inf, 
                                        'WIND_RATIO_BWI'].dropna(), 75)
andrewsmax = np.percentile(baywide_data.loc[baywide_data['WIND_RATIO_ANDREWS'] != np.inf, 
                                            'WIND_RATIO_ANDREWS'].dropna(), 75)
baywide_data['WIND_RATIO_BWI'] = baywide_data['WIND_RATIO_BWI'].replace(np.inf, bwimax)
baywide_data['WIND_RATIO_ANDREWS'] = baywide_data['WIND_RATIO_ANDREWS'].replace(np.inf, andrewsmax)
print(bwimax)
print(andrewsmax)
````

````python
def unpack_index(df, idvars_):
    df.index.name = 'old_index'
    df2 = df.reset_index()
    for c_ix, c_name in enumerate(idvars_):
        if c_name == 'Station':
            # the station levels are sorted into the correct order automatically 
            df2[c_name] = df2['old_index'].apply(lambda x: x[c_ix]).astype('category')
        else:
            df2[c_name] = df2['old_index'].apply(lambda x: x[c_ix])

    # order the data by date and then station
    df3 = df2.drop('old_index', axis=1).sort_values(by=idvars_)
    return df3.copy()

melt_df4 = unpack_index(melt_df2, id_vars)
print(melt_df4.columns)
````

````python
import pytz

def add_linear_time(pdsrs):
    my_timezone = pytz.timezone('US/Eastern')
    x = pdsrs.dt.tz_localize(my_timezone)
    winter_solstice = pd.to_datetime('12-21-2015 11:47pm').tz_localize(my_timezone)
    date1 = pd.to_datetime('01-01-2015 00:00').tz_localize(my_timezone)
    date2 = pd.to_datetime('01-01-2016 00:00').tz_localize(my_timezone)
    sec_per_yr = (date2-date1).total_seconds()
    return x.apply(lambda x: (x-winter_solstice).total_seconds()/sec_per_yr).astype(np.float64)

def add_day_length(lintime_srs):
    return lintime_srs.apply(lambda d: -1*np.cos(2*np.pi*(d))).astype(np.float64)

def add_fall_peak(lintime_srs):
    return lintime_srs.apply(lambda d: -1*np.sin(2*np.pi*(d))).astype(np.float64)

def add_spring_peak(lintime_srs):
    return lintime_srs.apply(lambda d: np.sin(2*np.pi*(d))).astype(np.float64)
````

````python
def cumulative_effector(bwdfx, date_col1, date_col2, to_agg, append_df):
    bwdf = bwdfx.copy().sort_values(by=date_col1)
    # make daily names
    for bwc in to_agg:
        print(bwc)
        for intvl, intvl_int in zip(['instant', '15', '30', '60'], [1, 15, 30, 60]):
            bwcInt = "-".join([bwc, intvl])
            append_df[bwcInt] = pd.Series(index=append_df.index)
            print(intvl)
            for ms_sd in append_df[date_col2].unique():
                row_selector = append_df[date_col2] == ms_sd
                ms_sd2 = ms_sd -  pd.to_timedelta(intvl_int, unit='d')
                row2_selector = (bwdf[date_col1] > ms_sd2) & (bwdf[date_col1] <= ms_sd)
                slice_to_recieve = bwdf.copy().loc[row2_selector, bwc]
                slice_to_recieve = slice_to_recieve.dropna()
                if (len(slice_to_recieve) == intvl_int):
                    append_df.loc[row_selector, bwcInt] = np.nansum(slice_to_recieve)
                elif (len(slice_to_recieve) > (0.2*intvl_int)):
                    append_df.loc[row_selector, bwcInt] = np.nanmean(slice_to_recieve)*len(slice_to_recieve)
                else:
                    append_df.loc[row_selector, bwcInt] = np.nan
                
    return append_df.copy()


bay_wide_columns = ['Discharge_Rappahannock', 'Discharge_Potomac', 'Discharge_Susquehanna', 
                    'interpolated_CHLA', 'measurement_PAR', 'WIND_RATIO_BWI', 'PRECIP24HR_BWI']

micro_df4 = cumulative_effector(bwdfx=baywide_data3, date_col1='DATE_TIME', date_col2='SampleDate',
                                to_agg=bay_wide_columns, append_df=micro_df3)
````

````python
import matplotlib.pyplot as plt
import scipy.stats as st

ixs_ = pd.IndexSlice

def make_time_label(c_id):
    left_lim = cb_limits[c_id][0]
    if left_lim[:2] == '01':
        return left_lim[:2]+left_lim[-5:]
    elif left_lim[:2] in ['06', '07', '08']:
        return left_lim[:5]
    else:
        return left_lim[:2]

color_codes = plt.rcParams['axes.prop_cycle'].by_key()['color']

def plot_time_series(predf, vars_, txfm, standardize, legend_loc, type_=[1,1], ylims=None):
    # ensures that the numerical index is ordered
    df = predf.reset_index()    
    # pulls out the raw data to plot in background
    y_srs = df[vars_]
    # converts them into a numpy array
    y_vals = y_srs.values
    # makes horizontal position a numerical series (offet by one from the index)
    x_vals = np.arange(1, 1+len(y_vals))
    # makes frame
    f, ax = plt.subplots(figsize=(8, 4), dpi=150)
    
    # create a label at the left of every new cruise start date 
    x_labels, x_tick_itvl, x_means = [], [], {}
    for c_id in df.cruise_id.unique():
        sub_cruise = df[df.cruise_id == c_id]
        x_labels.append(make_time_label(c_id))
        x_tick_itvl.append(x_vals[sub_cruise.index[0]])
        # positions the cruisly mean at the midpoint of all within cruise data
        x_means[c_id] = x_vals[sub_cruise.index[int(len(sub_cruise)/2)]]    
    
    # assign xticks to said intervals and rotated labels to ticks
    ax.xaxis.set_ticks(np.array(x_tick_itvl))
    _ = ax.set_xticklabels(x_labels)
    ax.xaxis.set_tick_params(rotation=270)
    # turns on grid
    ax.grid()
    
    # horizontal offsets for plotting multiple variables so error bars don't overlap
    offsets = np.arange(len(vars_))*2
    monthly_vals = {}
    for var_idx, var_n in enumerate(vars_):
        # pull out data fron a single variable
        spec_y = pd.Series(data=y_vals[:,var_idx], index=x_vals)
        
        # ensure no negative (raw) data (sometimes t-value distributions push CI into negative)
        # pulls out the desired variables and calculates cruisly means + CI 
        
        if type_[var_idx] == 1:
            def upperCI(x): 
                return st.t.interval(1-0.05, len(x.dropna())-1, loc=np.mean(x.dropna()), scale=st.sem(x.dropna()))[-1]
            def lowerCI(x):
                return st.t.interval(1-0.05, len(x.dropna())-1, loc=np.mean(x.dropna()), scale=st.sem(x.dropna()))[0]
            cruisly_means = df[[var_n, 'cruise_id']].groupby('cruise_id').agg([np.mean, upperCI, lowerCI])
        elif type_[var_idx] == 2:
            def upperCI(x): 
                return st.t.interval(1-0.05, len(x.dropna())-1, loc=np.sum(x.dropna()), scale=st.sem(x.dropna()))[-1]
            def lowerCI(x):
                return st.t.interval(1-0.05, len(x.dropna())-1, loc=np.sum(x.dropna()), scale=st.sem(x.dropna()))[0]
            cruisly_means = df[[var_n, 'cruise_id']].groupby('cruise_id').agg([np.sum, upperCI, lowerCI])
            
        monthly_vals[var_n] = cruisly_means.copy()
        # adds these positions this to the y-axis data
        cruisly_means['x'] = pd.Series(x_means)
        # drops cruises with all missing data (mid july lower bay)
        cruisly_means = cruisly_means.dropna()
        mean_y = cruisly_means.loc[:, ixs_[var_n, :]]
        
        # mask NAs
        spec_y = spec_y.dropna()
        # log transform raw data and means necessary
        if txfm:
            print('transforming hf data')
            if spec_y.min() < 0:    
                spec_y = np.log(spec_y + 1 + abs(spec_y.min()))
            else:
                spec_y = np.log(spec_y + 1)
            print('transforming cruise means')
            mean_y = mean_y.apply(lambda vv: np.log(vv + 1))
            
        # standard scale
        if standardize:
            print('standardizing hf data')
            spec_y_std = np.nanstd(spec_y)
            spec_y_mean = np.nanmean(spec_y)
            norm_y = (spec_y - spec_y_mean) / spec_y_std
            print('standardizing means')
            if type_[var_idx] == 1:
                mean_y_std = np.nanstd(mean_y.loc[:, ixs_[var_n, 'mean']])
                mean_y_mean = np.nanmean(mean_y.loc[:, ixs_[var_n, 'mean']])
                norm_mean_y = (mean_y - mean_y_mean) / mean_y_std
            elif type_[var_idx] == 2:
                mean_y_std = np.nanstd(mean_y.loc[:, ixs_[var_n, 'sum']])
                mean_y_mean = np.nanmean(mean_y.loc[:, ixs_[var_n, 'sum']])
                norm_mean_y = (mean_y - mean_y_mean) / mean_y_std
            # print mean & standard deviation for caption if necessary
            print("{}|Avg:{:.2f}|Std:{:.2f}".format(var_n, spec_y_mean, spec_y_std))
        else:
            norm_y = spec_y
            norm_mean_y = mean_y
            
        # plot as scatter
        ax.scatter(x=norm_y.index, y=norm_y.values, s=1.5, 
                   label=None, c=color_codes[var_idx], alpha=0.6)
        # convert CI into error 
        if type_[var_idx] == 1:
            asymmetric_error = [norm_mean_y.loc[:, ixs_[var_n, 'mean']] - norm_mean_y.loc[:, ixs_[var_n, 'lowerCI']], 
                                norm_mean_y.loc[:, ixs_[var_n, 'upperCI']] - norm_mean_y.loc[:, ixs_[var_n, 'mean']]]
            # plot cruisly means with CI
            ax.errorbar(cruisly_means.x+offsets[var_idx], norm_mean_y.loc[:, ixs_[var_n, 'mean']], 
                        yerr=asymmetric_error, marker='D', label=var_n, 
                        color=color_codes[var_idx], alpha=0.8)
        elif type_[var_idx] == 2:
            asymmetric_error = [norm_mean_y.loc[:, ixs_[var_n, 'sum']] - norm_mean_y.loc[:, ixs_[var_n, 'lowerCI']], 
                                norm_mean_y.loc[:, ixs_[var_n, 'upperCI']] - norm_mean_y.loc[:, ixs_[var_n, 'sum']]]
            # plot cruisly means with CI
            ax.errorbar(cruisly_means.x+offsets[var_idx], norm_mean_y.loc[:, ixs_[var_n, 'sum']], 
                        yerr=asymmetric_error, marker='D', label=var_n, 
                        color=color_codes[var_idx], alpha=0.8)
    if ylims:
        ax.set_ylim(bottom=min(ylims), top=max(ylims))
    ax.set_ylabel('Standard Deviations')
    ax.legend(loc=legend_loc)
    plt.show()
    return monthly_vals

````

````python
import statsmodels.api as sm
from statsmodels.formula.api import ols, rlm
from scipy.stats import gmean

model_error2 = pd.DataFrame(index=responses_set1+responses_set2, 
                            columns=predictor_sets1+predictor_sets2)
model_pvals2 = model_error2.copy()
counter = 0
countertotal = len(predictor_sets2+predictor_sets1)*len(responses_set1)
countertotal += len(predictor_sets2)*len(responses_set2)
print(countertotal)
countertotal *= 10

for pcol in predictor_sets1+predictor_sets2:
    for rcol in responses_set1+responses_set2:
        is_baywide_spatial = not ((rcol in responses_set2) and (pcol in predictor_sets1))
        is_not_unity = not ((pcol == rcol) or (pcol.split("-")[0] == rcol.split("-")[0]))
        if (is_baywide_spatial) and (is_not_unity):
            data = everything_df3.copy()[[pcol, rcol]].dropna()
            data.loc[data[rcol] < 0, rcol] = 0
            if rcol in log_cols:
                data[rcol] = np.log(data[rcol] + 1)
            if pcol in log_cols:
                data[pcol] = np.log(data[pcol] + 1)
            scale_col = lambda x: (x - x.mean())/(x.std())
            data2 = data.apply(scale_col, axis=0)
            data2['const'] = pd.Series([1.]*len(data2), index=data2.index)
            score_folds = pd.DataFrame(index=range(1,10), columns=['MAE', 'p-val'])
            for cv in range(1, 11):
                counter += 1 
                train_set = data2.sample(int(data2.shape[0]*0.75), random_state=cv)
                test_idx = set(data2.index) - set(train_set.index)
                print(pcol, rcol, len(train_set), len(test_idx), counter/countertotal)
                response_ = train_set[rcol]; 
                predictors_ = train_set[pcol];
                huber_t = sm.RLM(response_, predictors_, M=sm.robust.norms.HuberT(), missing='drop')
                hub_results = huber_t.fit()
                hub_test = hub_results.predict(exog=data2.loc[test_idx, pcol], transform=False)
                hub_resid = hub_test - data2.loc[test_idx, rcol]
                score_folds.loc[cv, 'MAE'] = np.mean(abs(hub_resid))
                score_folds.loc[cv, 'p-val'] = hub_results.pvalues.loc[pcol]
                
            model_error2.loc[rcol, pcol] = score_folds['MAE'].mean()
            model_pvals2.loc[rcol, pcol] = gmean(score_folds['p-val'].astype(np.float64))
````
